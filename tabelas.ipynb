```python
#resultados : uma tupla com (x,y,nome) de cada uma das linhas
def grafico_com_valores(resultados): 
    plt.figure(figsize=(10, 6))
    
    for resultado in resultados:
        x, y, nome = resultado
        plt.plot(x, y, label=f"{nome}", marker='o') 

        for xi, yi in zip(x,y):
            plt.text(xi, yi, f'{yi:.3f}', fontsize=9, ha='left', va='top')

        plt.grid(True)
        plt.title('Melhor modelo em cada faculdade')
        plt.xlabel('Número de batches')
        plt.ylabel('Acurácia')
        plt.legend()

    plt.show() 

```
